# Multivariate Regular Variation

Notebook to play around with different multivariate regular variation concepts and estimation techniques.

```r
library(mvtnorm)
library(dplyr)
library(ggplot2)

source('~/isolines_uq/scripts/R/auxiliary_scripts/distributionIsolines.R')
```

```r
n <- 100000
dat <- abs(data.frame(rmvt(n=n, sigma=matrix(c(1, 0.4, 0.4, 1), nrow=2), df=4)))
```

```r
# threshold selection for estimation
k <- 3000
```

```r
z1_k <- sort(dat$X1, decreasing=TRUE)[k]
z2_k <- sort(dat$X2, decreasing=TRUE)[k]
```

```r
norm_dat = data.frame(Z1 = dat$X1/z1_k, Z2 = dat$X2/z2_k)
```

```r
# lower lefthand corner of set
x <- 5
y <- 5

nu <- sum((norm_dat$Z1 > x) & (norm_dat$Z2 > y))/k
```

```r
# an estimate
nu*(k/n)
```

```r
((n/k))*x
```
