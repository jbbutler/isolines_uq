Work I'm doing to try to institute an adaptive $\beta_{n}$ procedure.

## Background

In Mammen and Polonik (2011), they prove the following sandwich bound on the probability of successfully covering your level set of interest.

\begin{align}
    (1-\alpha) - \mathbb{P}(A_{n}^{c}) + (p_{n}-F_{Z_{n}}(b_{n}^{+})) \leq \mathbb{P}[\hat{C}_{\ell} \subset C^{-} \text{ and } C \subset \hat{C}_{u}] \leq (1-\alpha) + (p_{n0}-F_{Z_{n}}(b_{n}^{-}))
\end{align}

where:
+ $A_{n} = \left\{ \sup_{x \in \mathbb{R}^{2}} |\hat{h}_{n}(x) - h(x)| \leq \beta_{n} \right\}$
+ $Z_{n}(\beta) = \sup_{x \in \mathbb{R}^{2}: |h(x)| \leq \beta} |\hat{h}_{n}(x) - h(x)|$
+ $p_{n} = \mathbb{P}[Z_{n}(\beta) \leq \hat{b}_{n}]$
+ $p_{n0} = \mathbb{P}[Z_{n}(0) \leq \hat{b}_{n}]$
+ $p_{n}^{-} = \mathbb{P}[Z_{n}(\beta) \leq b_{n}^{-}]$
+ $p_{n}^{+} = \mathbb{P}[Z_{n}(\beta) \leq b_{n}^{+}]$
+ $b_{n}^{-} = \inf\{t \in \mathbb{R} : F_{Z_{n}}(t) \geq 1-\alpha\}$, $b_{n}^{+} = \sup\{t \in \mathbb{R} : F_{Z_{n}}(t) \leq 1-\alpha\}$
