# DKW Results

Using the DKW inequality to derive $c_{1-\alpha}$, in the confidence tube construction $\hat{C} = \{x \in \mathbb{R}^{2}: p-c_{1-\alpha} \leq \hat{\bar{F}}(x) \leq p + c_{1-\alpha}\}$

## Bivariate t

```r
library(dplyr)
library(ggplot2)

path <- '/pscratch/sd/j/jbbutler/dkw_analysis/bivt_dkw_results.RData'
res_df <- readRDS(path)
```

```r
res_df %>% group_by(alpha, p, n) %>% summarize(coverage_rate=mean(covered))
```

Theoretically, using DKW in this way gives us a confidence tube construction with guaranteed coverage at least $1-\alpha$, so it gives us at worst conservative confidence tubes, but at least we never undercover.
