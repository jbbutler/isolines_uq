# Asymptotic Misspecification

How badly can you screw up if you run the method assuming asymptotic dependence when the data is actually asymptotically independent, and vice versa?

```r
# load libraries/source necessary functions
library(ggplot2)
library(dplyr)
library(mvtnorm)
library(gridExtra)
library(repr)

source('/global/u1/j/jbbutler/isolines_uq/scripts/R/orig_isolines.R')
```

#### Assume Asymptotically Independent, but Actually Asymptotically Dependent

We take the asymptotically independent case to be data generated from a bivariate gaussian, and the asymptotically dependent case to be data generated from a bivariate t.

```r
n <- 1000
dat <- data.frame(rmvt(n, sigma = matrix(c(1, 0.7, 0.7, 1), nrow = 2), df = 4))

out_wrong <- xContours(dat = dat, asympIndep = TRUE, faster = TRUE)
out_correct <- xContours(dat = dat, asympIndep = FALSE, faster = TRUE)
wrong_1 <- data.frame(out_wrong$contourOrig)
correct_1 <- data.frame(out_correct$contourOrig)
wrong_2 <- data.frame(out_wrong$projContours[[1]])
correct_2 <- data.frame(out_correct$projContours[[1]])
wrong_3 <- data.frame(out_wrong$projContours[[2]])
correct_3 <- data.frame(out_correct$projContours[[2]])
wrong_4 <- data.frame(out_wrong$projContours[[3]])
correct_4 <- data.frame(out_correct$projContours[[3]])

n_1000 <- ggplot() +
    geom_point(data = dat, aes(x = X1, y = X2), alpha = 0.1) + 
    geom_line(data = wrong_1, aes(x = X1, y = X2), col = 'red') + 
    geom_line(data = wrong_2, aes(x = X1, y = X2), col = 'red') + 
    geom_line(data = wrong_3, aes(x = X1, y = X2), col = 'red') + 
    geom_line(data = wrong_4, aes(x = X1, y = X2), col = 'red') + 
    geom_line(data = correct_1, aes(x = X1, y = X2), col = 'green') +
    geom_line(data = correct_2, aes(x = X1, y = X2), col = 'green') +
    geom_line(data = correct_3, aes(x = X1, y = X2), col = 'green') +
    geom_line(data = correct_4, aes(x = X1, y = X2), col = 'green') +
    xlim(0, NA) +
    ylim(0, NA) +
    ggtitle(paste('n =', n)) +
    theme_light()
```
